# Timing Battery Example

Run the battery directly from Python. Saving outputs is optional and disabled by default.

## Setup Imports

```python
from pathlib import Path
import sys

# Set this to your gbnet repository path when running this notebook elsewhere.
GBNET_REPO_ROOT = Path("/Users/michaelhorrell/github/gbnet").expanduser().resolve()
assert (GBNET_REPO_ROOT / "scripts" / "timing_battery.py").exists()

if str(GBNET_REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(GBNET_REPO_ROOT))

from scripts.timing_battery import (
    BatteryConfig,
    DISPLAY_COLUMNS,
    run_battery_and_record,
)
```

## Run Battery

```python
SAVE_OUTPUTS = False

history_path = GBNET_REPO_ROOT / "benchmarks" / "timing_history.json"
run_json_path = GBNET_REPO_ROOT / "benchmarks" / "timing_latest.json"

config = BatteryConfig(
    n_samples=3000,
    n_features=32,
    output_dim=1,
    n_rounds=20,
    repeats=3,
    predict_passes=50,
    seed=2026,
    num_threads=1,
    gblinear_lr=0.5,
    gblinear_lambd=0.01,
)

timing_df, payload = run_battery_and_record(
    config=config,
    xgb_params={},
    lgb_params={},
    save_outputs=SAVE_OUTPUTS,
    history_path=history_path,
    run_json_path=run_json_path,
    include_git_metadata=SAVE_OUTPUTS,
    repo_root=GBNET_REPO_ROOT,
)

print("Run key:", payload["run_key"])
print("Outputs saved:", SAVE_OUTPUTS)
```

## Current Timing DataFrame

```python
timing_df[DISPLAY_COLUMNS]
```

## Optional Saved History

```python
import json
import pandas as pd

if SAVE_OUTPUTS:
    history = json.loads(history_path.read_text())
    history_df = pd.DataFrame(history["runs"]).T
    history_df[["run_datetime_utc", "commit", "run_key"]].tail(5)
else:
    print("SAVE_OUTPUTS is False, so no history was written.")
```
